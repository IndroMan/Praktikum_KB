```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("data-item.csv")
```

```python
df
```

<h3> Menggunakan Visualisasi Bar <br>
    Disini saya menampilkan data atribut kategori dan totalReviews <br>
    Untuk Melihat berapa banyak review pada kategori tersebut

```python
x = df[('category')]
y = df[('totalReviews')]
```

```python
plt.barh(x, y, align="edge")
plt.title("Total Review")
plt.xlabel("Jumlah")
plt.ylabel("Kategori")
plt.show()
```

<h3> Visualisasi selanjutnya saya menggunakan grafik pie <br>
    disini saya akan memperlihatkan banyak nya review di setiap kategori

```python
df['category'].value_counts()
```

```python
labels = df['category'].value_counts().rename_axis('category').reset_index(name='Jumlah')
labels
```

```python
plt.figure(figsize=(8,8), dpi=200)
plt.pie(df['category'].value_counts(), labels=labels['category'], autopct='%.1f%%', startangle=90)
plt.show()
```

<h3> Grafik Scatter Plot <br>
    diagram grafis yang dibangun dari dua sumbu X dan Y (variabel X dan variabel Y). Scatter plot menunjukan hubungan antara dua variabel.

Implementasi grafik Scatter Plot sebagai berikut

```python
sns.scatterplot(data = df, x='category', y = 'totalReviews')
plt.title("Persebaran data ukuran totalReviews dengan category yang beda")
plt.show()
```

<h3> Grafik Heatmap <br>
    grafik ini digunakan untuk melihat Korelasi antar atribut.

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), cmap='BuPu')
ax.set_title('Korelasi antar atribut pada dataset')
plt.show()
```

 <h5>Dari tampilan grafik diatas menyatakan <br>
tidak ada korelasi yang baik antar atribut yang berbeda
